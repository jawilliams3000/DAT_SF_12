####8. Which type of complaint was most common?

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE
df2 = pd.read_sql("""

        SELECT a.neighborhood, a.request_type, a.num FROM
        (SELECT 
            neighborhood, request_type, Count(1) AS num
        FROM event 
        GROUP BY neighborhood, request_type
        ) a
        INNER JOIN
        (
        SELECT neighborhood, MAX(num) AS maxNum 
        FROM
            (
                SELECT 
                    neighborhood, request_type, Count(1) AS num
                FROM event 
                GROUP BY neighborhood, request_type
            ) 
            GROUP BY neighborhood
        ) b
        ON a.neighborhood = b.neighborhood
            AND a.num = b.maxNum
        WHERE a.request_type != 'Sidewalk_Cleaning'
        ORDER BY num DESC
    """, con)
df2

```

```python
#YOUR CODE HERE
#What is the most common category complaint per supervisor district ?
df2 = pd.read_sql("""

        SELECT a.supervisor_district, a.category, a.num FROM
        (SELECT 
            supervisor_district, category, Count(1) AS num
        FROM event 
        GROUP BY supervisor_district, category
        ) a
        INNER JOIN
        (
        SELECT supervisor_district, MAX(num) AS maxNum 
        FROM
            (
                SELECT 
                    supervisor_district, category, Count(1) AS num
                FROM event 
                GROUP BY supervisor_district, category
            ) 
            GROUP BY supervisor_district
        ) b
        ON a.supervisor_district = b.supervisor_district
            AND a.num = b.maxNum
        ORDER BY a.supervisor_district ASC
    """, con)
df2
```
