```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from seaborn import plt, color_palette
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv('..\data\wine.csv')
```

```python
wine.describe()
```

```python
sns.pairplot(wine, size=1.4, aspect=1.4)
```

```python
# Various variables we'll need to set intially.
n_neighbors = range(1, 51, 2)
```

```python
wine_variables = wine.drop('Wine',axis=1)
print wine_variables
```

```python
# Create the training (and test) set using scikit-learn's train_test_split function
X_train, X_test, y_train, y_test = train_test_split(wine_variables, wine['Wine'], test_size=0.3, random_state=42)
```

```python
# Loop through each neighbors value from 1 to 51 and append
# the scores
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```

```python
plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
```

```python
from sklearn.cross_validation import cross_val_score
clf = neighbors.KNeighborsClassifier(27, weights='uniform')
scores = cross_val_score(clf, X_test, y_test, cv=5)
print scores
```

```python
np.mean(scores)
```

```python
# See what features return the highest significance
n = np.arange(len(wine_variables.columns.values))

fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot(111)

ax.bar(n, feature_selection.f_classif(X_test, y_test)[0])

xtickNames = ax.set_xticklabels(wine_variables.columns.values)
ax.set_xticks(n)
_ = plt.setp(xtickNames, rotation=45, fontsize=16)
```

Note that the Wine group codes will not match between Clustering and Wine Category. What we are looking for is does all the wines in the cluster fall into the same Wine Category).

```python
#drop first column
wine2 = wine[['Wine','Flavanoids','Proline']]
```

```python
wine2.describe()
```

```python
plt.scatter(wine2['Flavanoids'], wine2['Proline']);
```

```python
#c = [color_palette()[i] for i in wine['Wine']]
#plt.scatter(wine2[:,0], wine2[:,1], c=c)
```

```python
# compute distance matrix
from scipy.spatial.distance import pdist, squareform

# not printed as pretty, but the values are correct
distx = squareform(pdist(wine2, metric='euclidean'))
distx
```

We then use scipy.cluster.hierarchy.linkage to create the hierarchy, and ...dendrogram to plot it.

```python
# perform clustering and plot the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

R = dendrogram(linkage(distx, method='single'), color_threshold=10)

plt.xlabel('Flavanoids')
plt.ylabel('Proline')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
R = dendrogram(linkage(distx, method='ward'), color_threshold=100)

plt.xlabel('Flavanoids')
plt.ylabel('Proline')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarch = AgglomerativeClustering(n_clusters=3, linkage='ward')
```

```python
Y_hat_hier = hierarch.fit(wine2).labels_
```

```python
c_hat = [color_palette()[i] for i in Y_hat_hier]
plt.scatter(wine2['Flavanoids'], wine2['Proline'], c=c_hat);
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(3, random_state=1)
Y_hat_kmeans = kmeans.fit(wine2).labels_

c_hat = [color_palette()[i] for i in Y_hat_kmeans]

plt.scatter(wine2['Flavanoids'], wine2['Proline'], c=c_hat);
```

```python
plt.scatter(wine2['Flavanoids'], wine2['Proline'], c=c_hat, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=[color_palette()[i] for i in np.unique(Y_hat_kmeans)])
print mu
```

```python
wine_predict = pd.DataFrame(Y_hat_kmeans)
```

```python
wine_compare = pd.merge(wine2, wine_predict, left_index=False, right_index=False, how='outer')
```
